Pandas Profiling

```python
pip install ydata-profiling pandas
```

```python
import pandas as pd
from ydata_profiling import ProfileReport

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Generate EDA report
profile = ProfileReport(df, explorative=True)

# Save the report as an HTML file
profile.to_file("titanic_report.html")

# Display the report inside Jupyter Notebook (if using one)
profile.to_notebook_iframe()
```

```python
# Import necessary libraries
from autoviz.AutoViz_Class import AutoViz_Class
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Create an AutoViz instance
AV = AutoViz_Class()

# Perform AutoViz analysis, call the class instance with appropriate parameters
dft = AV.AutoViz(filename="", dfte=df) # The method name is actually the class name itself, 'AutoViz'
```
